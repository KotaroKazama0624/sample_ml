```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#データの読み込み、整形
df = pd.read_csv('nikkei_stock_average_daily_jp.csv', encoding="shift-jis")
df = df[:-1]
df.columns = ['Date', 'Close', 'Open', 'High','Low']
df
```

```python
df.info()
```

```python
#データの型をdatetime型へ変更
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
#曜日情報を追加(月曜:0, 火曜:1, 水曜:2, 木曜:3, 金曜:4、土曜:5、日曜:6)
df['weekday'] = df['Date'].dt.weekday
#日時データを加工しやすいように2019-01-04のデータを削除
df = df.drop(0)
df
```

```python
#初めの月曜日となる2019/1/7を基準に日数を追加
from datetime import datetime
from datetime import timedelta
start = datetime(2019,1,7)
df['weeks'] = (df['Date'] - start) // timedelta(weeks=1)
df.head(10)
```

```python
#分かりやすいようにカラムの並び替え
df = df[['Date', 'weeks', 'weekday', 'High', 'Low', 'Open', 'Close']]
df
```

```python
#データの並び替え
df.sort_values(by='Date', ascending=True, inplace=True)

#日付をindexにセット
df.set_index(keys='Date', inplace=True)
df
```

```python
df_shift = df.shift(-1)
df_shift
```

```python
#翌日の終値と本日の終値の差分を追加
df['delta_Close'] = df_shift['Close'] - df['Close']
df
```

```python
#目的変数Upを追加する（翌日の終値が上がる場合は1, それ以外は0とする）、'delta_Close'カラムの削除
df['Up'] = 0
df['Up'][df['delta_Close'] > 0] = 1
df = df.drop('delta_Close', axis=1)
df
```

```python
#グラフ化のためにカラムを抽出
df_new = df[['Close', 'Open', 'High', 'Low']]
df_new.plot(kind='line')
plt.show()
```

```python
#終値の前日比の追加
df_shift = df.shift(1)
df['Close_ratio'] = (df['Close'] - df_shift['Close']) / df_shift['Close']
df
```

```python
# 始値と終値の差分を追加
df['Body'] = df['Open'] - df['Close']
df
```

```python
# 週番号をリストに格納
list_weeks = []
list_weeks = df['weeks'].unique()
list_weeks
```

```python
# 各週ごとの日数を入力
df['week_days'] = 0

for i in list_weeks:
    df['week_days'][df['weeks']==i] = len(df[df['weeks'] == i])

df
```

```python
#月曜～金曜まで5日分データのある週だけデータを取り出す
df = df[df['week_days'] == 5]
df
```

```python
#金曜日のデータを削除する(weekday:4となるデータ)
df = df[df['weekday'] != 4]
df
```

```python
#不要カラムの削除と並び替え
df = df[['weekday', 'High', 'Low', 'Open', 'Close', 'Close_ratio', 'Body', 'Up']]
df
```

```python
# 学習データを2019-01-07～2020-12-31の期間としdf_trainに入力する
df_train = df['2019-01-07' : '2020-12-31']
df_train
```

```python
#検証データを2021-01-01以降としてdf_valに入力する
df_val = df['2021-01-01' : ]
df_val
```

```python
# 学習データを説明変数(X_train)と目的変数(y_train)に分ける(LSTM用)
X_train = df_train[['weekday', 'High', 'Low', 'Open', 'Close', 'Close_ratio', 'Body']]
y_train = df_train['Up']
# 学習データを説明変数(X_train)と目的変数(y_train)に分ける(SVM用)
# schikit-learnのsvmは2次元データしか扱えないので時系列分析で扱うために説明変数を終値だけにする
X_train_for_svm = df_train['Close']
y_train_for_svm = df_train['Up']
#確認
print(X_train)
print(y_train)
```

```python
# 検証データを説明変数(X_val)と目的変数(y_val)に分ける(LSTM用)
X_val = df_val[['weekday', 'High', 'Low', 'Open', 'Close', 'Close_ratio', 'Body']]
y_val = df_val['Up']
# 検証データを説明変数(X_val)と目的変数(y_val)に分ける(SVM用)
X_val_for_svm = df_val['Close']
y_val_for_svm = df_val['Up']
#確認
print(X_val)
print(y_val)
```

```python
# 学習データと検証データの終値（Close）の折れ線グラフ作成
X_train['Close'].plot(kind='line')
X_val['Close'].plot(kind='line')
plt.legend(['X_train', 'X_val'])
plt.show()
```

```python
# 標準化を行う
# StandardScalerのインポート
from sklearn.preprocessing import StandardScaler

#4日ごとにデータを抜き出して、標準化とnumpy配列に変換する関数（std_to_np）の定義
def std_to_np(df):
    df_list = []
    df = np.array(df)
    for i in range(0, len(df) - 3, 4):
        df_s = df[i:i+4]
        std = StandardScaler()
        #fit_transformでよい？基準がそれぞれのデータで違くなる？
        df_std = std.fit_transform(df_s)
        df_list.append(df_std)
    return np.array(df_list)

#svm用
def to_np(df):
    df_list = []
    df = np.array(df)
    for i in range(0, len(df) - 3, 4):
        df_s = df[i:i+4]
        df_list.append(df_s)
    return np.array(df_list)
```

```python
#学習データと検証データの説明変数にstd_to_npを実行（LSTM用）
X_train_np_array = std_to_np(X_train)
X_val_np_array = std_to_np(X_val)

#学習データと検証データの説明変数にto_npを実行（SVM用）
X_train_np_array_for_svm = to_np(X_train_for_svm)
X_val_np_array_for_svm = to_np(X_val_for_svm)

#学習データと検証データの形の確認
print(X_train_np_array_for_svm)
print(X_val_np_array_for_svm)
```

```python
#学習データと検証データの目的変数を確認
print(y_train)
print(y_val)
```

```python
# 学習データ、検証データの目的変数の間引き
# 週の４日目（木曜日）のデータだけ抜き出す
y_train_new = y_train[3::4]
y_val_new = y_val[3::4]

#間引きの学習データと検証データの目的変数を確認
print(y_train_new)
print(y_val_new)
```

```python
#時系列分割のためTimeSeriesSplitとGridsearchCVのインポート
from sklearn.model_selection import TimeSeriesSplit,GridSearchCV
#accuracy算出のためaccuracy_scoreのインポート
from sklearn.metrics import accuracy_score
```

```python
print(X_train_np_array)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
pipe_svc = make_pipeline(StandardScaler(),SVC(random_state=1))
param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
param_grid = [{'svc__C': param_range, 'svc__kernel': ['linear']}]
#              {'svc__C': param_range, 'svc__gamma': param_range,
#               'svc__kernel': ['rbf']}]
tscv = TimeSeriesSplit(n_splits=4).split(X_train_np_array_for_svm)
model = GridSearchCV(estimator=pipe_svc,
                    cv=tscv,
                    param_grid=param_grid,
                    scoring='accuracy',
                    refit=True,
                    n_jobs=-1)
model = model.fit(X_train_np_array_for_svm, y_train_new)
print(model.best_score_)
print(model.best_params_)
```

```python
#予測
y_valid_pred = model.predict(X_val_np_array_for_svm)
print('y pred', y_valid_pred)
# 予測精度の算出と表示
score = accuracy_score(y_val_new, y_valid_pred)
print('score: ',score)
#データをcsvで吐き出す
#df.to_csv('nikkei_stock_ave_shaped.csv')
```

```python
import pickle

#モデルの保存
filename = 'stockmodel.pkl'
#pickle.dump(model.best_estimator_, open(filename, "wb"))
#モデルの読み込み
loaded_model = pickle.load(open(filename, "rb"))
y_valid_pred_loaded = loaded_model.predict(X_val_np_array_for_svm)
print('y pred', y_valid_pred_loaded)
#読み込んだモデルのスコアを確認
score = accuracy_score(y_val_new, y_valid_pred_loaded)
print('score: ',score)
```
